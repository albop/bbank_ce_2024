## Simulating an AR1

Take an AR1 process $x_t = A  x_{t-1} + \epsilon_t$ with
$\epsilon_t \sim \Sigma$ where $\Sigma$ is a positive definite matrix.

<span class="theorem-title">**Exercise 12**</span> Define 2x2 matrices
$A$ and $\Sigma$, the latter being symmetric positive definite

```python
A = np.array([
    [0.9, 0.4],
    [-0.1, 0.8]
])
Σ = np.array([
    [0.8, 0.1],
    [0.1, 0.6]
])
```

<span class="theorem-title">**Exercise 13**</span> Compute asymptotic
variance using matrix algebra (there is a recursive formula)

Define $S_0 = I$ and define $S_{n+1} = A S_n A' + \Sigma$

```python
def asymptotic_variance(A,Σ, K=1000):
    n = Σ.shape[0]
    S = np.eye(n)
    for k in range(K):
        S = A@S@A.T + Σ
    return S
```

```python
asymptotic_variance(A,Σ)
```

<span class="theorem-title">**Exercise 14**</span> Simulate $N$ draws
for $T$ periods and store the result in `sim`.

```python
N = 50 # let's simulate N draws at the same time
def simulate( A, Σ, T, N):
    x0= np.array([0,0]) # initial point
    x = x0[None,:].repeat(N, axis=0)
    mu = Σ[0,:]*0
    sim = [x]
    for t in range(T):
        x = sim[-1]
        epsilon = np.random.multivariate_normal(mu, Σ, size=N)
        x = x@A.T + epsilon
        sim.append(x)
    # return sim
    # print(sim[0])
    return np.concatenate(
        [u[:,None,:] for u in sim],
        axis=1
    )
```

```python
sims = simulate(A, Σ, 100, 10)
sims.shape
# first dimension: draw
# second dimension: time
# third dimension: variable
```

<span class="theorem-title">**Exercise 15**</span> Compute ergodic
variance (bonus compute std of the variance estimate)

```python
sims = simulate(A, Σ, 100, 1000)
```

```python
np.cov(
    sims[:, -1, :].T # keep only last period
)
#orders of magnitude are similar to theoretical variance computed above
```
